# Importing Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

# Uploading Dataset

```python
df=pd.read_csv('breast_data.csv')
```

# First & Last 5 rows

```python
df.head(5)
```

```python
df.tail(5)
```

# Information of Dataset

```python
df.info()
```

# Dropping Unnecessary Column

```python
df=df.drop(columns="Unnamed: 32")
```

```python
df.head()
```

# Checking Statistical Data

```python
df.describe()
```

# Checking Null Value

```python
df.isnull().sum()
```

# Label Encoding(Transforming Categorical to Numerical Data)

```python
df['diagnosis'].value_counts()
```

```python
df["diagnosis"]=df["diagnosis"].astype('category').cat.codes
df
```

# Dividing the dataset into X & Y

```python
x=df.drop(columns=["diagnosis","id"])
y=df["diagnosis"]
```

# Apply Feature Scaling

```python
st=StandardScaler()
ss=st.fit_transform(x)
```

# Dividing the dataset into Train & Test

```python
x_train,x_test,y_train,y_test=train_test_split(ss,y,test_size=0.2,random_state=2)
```

# Training Model

# 1.Logistic Regression

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

**Accuracy of Training Data**

```python
x_train_pred=lr.predict(x_train)
accu_score=accuracy_score(x_train_pred,y_train)
print("Accuracy:",accu_score)
```

**Accuracy of Testing Data**

```python
x_test_pred=lr.predict(x_test)
accu_score=accuracy_score(x_test_pred,y_test)
print("Accuracy:",accu_score)
```

# 2. K Neighbourhood Classifier

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

**Accuracy of Trainging Data**

```python
x_train_pred=knn.predict(x_train)
accu_score=accuracy_score(x_train_pred,y_train)
print("Accuracy :",accu_score)
```

**Accuracy of Testing Data**

```python
x_test_pred=knn.predict(x_test)
accu_score=accuracy_score(x_test_pred,y_test)
print("Accuracy :",accu_score)
```

# 3.Decision Tree

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

**Accuracy of Training Data**

```python
x_train_pred=dt.predict(x_train)
accu_score=accuracy_score(x_train_pred,y_train)
print("Accuracy :",accu_score)
```

**Accuracy of Testing Data**

```python
x_test_pred=dt.predict(x_test)
accu_score=accuracy_score(x_test_pred,y_test)
print("Accuracy :",accu_score)
```

# 4.Random Forest

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

**Accuracy of Training Data**

```python
x_train_pred=rf.predict(x_train)
accu_score=accuracy_score(x_train_pred,y_train)
print("Accuracy :",accu_score)
```

**Accuracy of Testing Data**

```python
x_test_pred=rf.predict(x_test)
accu_score=accuracy_score(x_test_pred,y_test)
print("Accuracy :",accu_score)
```

# 5.Support Vector Machine

```python
svc=SVC()
svc.fit(x_train,y_train)
```

**Accuracy of Training Data**

```python
x_train_pred=svc.predict(x_train)
accu_score=accuracy_score(x_train_pred,y_train)
print("Accuracy :",accu_score)
```

**Accuracy of Testing Data**

```python
x_test_pred=svc.predict(x_test)
accu_score=accuracy_score(x_test_pred,y_test)
print("Accuracy :",accu_score)
```
